## PROBLEM 1

### Input:
อ่านพิกัดเชิงขั้วของจุดบนระนาบ ซึ่งเป็นจำนวนจริงสองตัวคั่นด้วยช่องว่าง เก็บไว้ในตัวแปร ```r``` และ ```theta``` (หน่วยเป็นเรเดียน)

### Process:
คำนวณค่า ```x``` และ ```y``` ซึ่งเป็นพิกัดคาร์ทีเซียน ของจุด ```(r, theta)``` ที่อ่านเข้ามา

$$\begin{align}
x &= r\cos(\theta), \\
y &= r\sin(\theta)
\end{align}$$

### Output:
ค่า ```x``` และ ```y``` คั่นด้วยช่องว่าง

----------------------------------------------
## PROBLEM 2

### Input:
อ่านพิกัดคาร์ทีเซียนของจุดบนระนาบ ซึ่งเป็นจำนวนจริงสองตัวคั่นด้วยช่องว่าง เก็บไว้ในตัวแปร ```x``` และ ```y``` 

### Process:
คำนวณค่า ```r``` และ ```theta``` ซึ่งเป็นพิกัดบนระนาบเชิงขั้ว ของจุด ```(x, y)``` ที่อ่านเข้ามา

$$\begin{align}
r &= \sqrt{x^2 + y^2}, \\
y &= \arctan{\left(\dfrac{y}{x}\right)}
\end{align}$$

### Output:
ค่า ```r``` และ ```theta``` คั่นด้วยช่องว่าง

## PROBLEM 4

จงเขียนโปรแกรมคำนวณพื้นที่สามเหลี่ยมที่ทราบความยาวด้านสองด้าน ( $a$ กับ $b$) และมุมระหว่างด้านสองด้านนั้น ($c$) 

<img src="../images/tr1.png" alt="drawing" width="200"/>

จากสูตร
$$ \textrm{area} =  \dfrac{1}{2}ab\sin(c) $$


### ข้อมูลนำเข้า

- บรรทัดแรกคือความยาวด้าน $a$ (หน่วยเป็นเซนติเมตร)
- บรรทัดที่สองคือความยาวด้าน $b$ (หน่วยเป็นเซนติเมตร)
- บรรทัดที่สามคือมุมระหว่างด้านทั้งสอง $c$ (หน่วยเป็นองศา)


### ข้อมูลส่งออก

พื้นที่ของสามเหลี่ยมที่รับเป็นข้อมูลนำเข้า (หน่วยเป็นตารางเซนติเมตร) แสดงในรูปแบบที่แสดงตามตัวอย่างด้านล่าง 

### ตัวอย่าง 

| Input (แป้นพิมพ์)   |  Output (จอภาพ) |
|-----------------|------------------|
| 10<br>10<br>90 | area = 50.0 (sq.m)<br><br>|
| 10<br>20<br>50.5 | area = 77.1624 (sq.m)<br><br>|

## PROBLEM 5

ISBN เป็นตัวเลขจำนวน 10-13 หลักที่ใช้ระบุหนังสือแต่ละเล่ม โจทย์ข้อนี้ สนใจเฉพาะ ISBN ที่มี 10 หลัก การตรวจสอบความถูกต้องของ ISBN จะใช้ตัวเลขหลักสุดท้ายเป็น check digit ในการตรวจสอบ ความถูกต้องของตัวเลขอื่น ๆ โดยวิธีที่ใช้ตรวจสอบคือ 

$$ 10n_1 + 9n_2 + 8n_3 + 7n_4 + 6n_5 + 5n_6 + 4n_7 + 3n_8 + 2n_9 + n_{10}  $$

จะต้องถูกหารด้วย 11 ลงตัว

----------
ตัวอย่างเช่น หากตัวเลข 9 หลักแรกคือ ***020131452*** จะไต้ว่า


$$ 10*0 + 9*2 + 8*0 + 7*1 +6*3 + 5*1 +4*4 + 3*5 + 2*2 + n_{10} =83 + n_{10} $$ 


ต้องถูกหารด้วย 11 ลงตัว

จะไต้ว่า $n_{10}$ ต้องมีค่าเท่ากับ 5 เพื่อให้ผลรวมเป็น 88 ซึ่งหารด้วย 11 ลงตัว และไต้ ISBN คือ ***0201314525***

----------

หากกำหนดตัวเลขหลักที่ 1-9 มาให้ จงคำนวณหา ISBN ทั้งสิบหลัก 


### ข้อมูลนำเข้า

มีบรรทัดเดียว ระบุ ISBN หลักที่ 1-9


### ข้อมูลส่งออก

มีบรรทัดเดียว แสดง ISBN ทั้งสิบหลัก

### ตัวอย่าง 

| Input (แป้นพิมพ์)   |  Output (จอภาพ) |
|-----------------|------------------|
| 020131452 | 0201314525  |
| 100000000 | 1000000001  |
